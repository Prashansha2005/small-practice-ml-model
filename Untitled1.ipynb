```python
df = df.dropna(subset=['country'])
df['is_us'] = (df['country'] == 'us').astype(int)

```

```python
x = df.iloc[:, [5,10]]

y=df.iloc[:,-1]
x = x.apply(pd.to_numeric, errors='coerce')
y = pd.to_numeric(y, errors='coerce')

# Drop any rows that became NaN after conversion
mask = ~x.isna().any(axis=1) & ~y.isna()
x = x[mask]
y = y[mask]
```

```python
from sklearn.model_selection import train_test_split
# Drop rows where latitude is NaN


x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.1)
x_train = x_train.dropna()
x_test = x_test.dropna()
y_train = y_train.dropna()
y_test = y_test.dropna()
```

```python
x_train
```

```python
# Combine X and y temporarily
df_combined = x_train.copy()
df_combined['is_us'] = y_train  # temporarily hold y

# Drop rows where any column (or target) has NaN
df_combined = df_combined.dropna(subset=[x_train.columns[0], x_train.columns[1], 'is_us'])

# Split back into X and y
x_train_clean = df_combined[x_train.columns]
y_train_clean = df_combined['is_us']


# Combine X and y temporarily
df_combinedtes = x_test.copy()
df_combinedtes['is_us'] = y_test  # temporarily hold y

# Drop rows where any column (or target) has NaN
df_combinedtes = df_combinedtes.dropna(subset=[x_train.columns[0], x_train.columns[1], 'is_us'])

# Split back into X and y
x_test_clean = df_combinedtes[x_train.columns]
y_test_clean = df_combinedtes['is_us']


```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_clean=scaler.fit_transform(x_train_clean)
x_test_clean=scaler.fit_transform(x_test_clean)
```

```python
x_train_clean
```

```python
x_test_clean=scaler.fit_transform(x_test_clean)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train_clean,y_train_clean)
```

```python
y_pred=model.predict(x_test_clean)
```

```python
y_test_clean
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test_clean,y_pred)
print(f"accuracy:{acc}%")
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_test_clean,y_pred,clf=model,legend=2)
```
